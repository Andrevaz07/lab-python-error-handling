```python
# Step 1: Define the function for initializing the inventory with error handling
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory
```

````python
def calculate_total_price(customer_orders):
    total_price = 0.0
    
    for product in customer_orders:
        while True:
            try:
                price = float(input(f"Enter the price of the {product}: "))
                if price < 0:
                    raise ValueError("Price cannot be negative. Please enter a valid price.")
                
                total_price += price
                break  # Exit the loop if a valid price is entered
            except ValueError as ve:
                print(f"Error: {ve}")
    
    return total_price

# Example usage:
customer_orders = ["item1", "item2", "item3"]
total_price = calculate_total_price(customer_orders)
print(f"Total price: {total_price}€")




def get_customer_orders(inventory):
    while True:
        try:
            number_ordered = int(input("Enter the number of customer orders: "))
            if number_ordered < 0:
                raise ValueError("Number of orders cannot be negative. Please enter a valid number.")
            
            customer_orders = set()
            for i in range(number_ordered):
                product_name = input("What's the name of the product that a customer wants to order: ")
                
                if product_name not in inventory or inventory[product_name] == 0:
                    raise ValueError(f"Invalid product name or out of stock. Please enter a valid product name.")
                
                customer_orders.add(product_name)
            
            return customer_orders
        except ValueError as ve:
            print(f"Error: {ve}")

# Example usage with an inventory dictionary:
inventory = {"item1": 5, "item2": 10, "item3": 0}
customer_orders = get_customer_orders(inventory)
print("Customer Orders:", customer_orders)

````
